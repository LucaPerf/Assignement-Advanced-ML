```python
import io 
import numpy as np 
import random

import tensorflow as tf
from tensorflow import keras
from keras.callbacks import LambdaCallback
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

import re
```

```python
print('Opening file...')
path = "divina_commedia.txt"
with io.open(path, 'r', encoding='utf-8') as file:
    text = file.read().lower()
print("text length", len(text))
print()
print('\n\n\n\n\n', '***** first 1000 characters *****', '\n\n\n\n\n')
text[0:1000]
```

```python
# Rimuovere i salti di riga multipli per uniformare il formato
text = re.sub(r'\n+', '\n', text)  # Rimuove salti di riga multipli

canti = re.split(r'\n*(inferno|purgatorio|paradiso):\s*canto\s*[ivxlcdm]+\n*', text)
print(canti)
print(len(canti))
print(canti[4])
```

```python
# Rimuovere eventuali testo introduttivo o righe extra (come informazioni sull'autore)
re.sub(r"La Divina Commedia.*?Progetto Manuzio", "", text, flags=re.DOTALL)

with open("nuovo_divina.txt", "w", encoding="utf-8") as new_file:
    new_file.write(text)

# Regex per separare i canti (inferno, purgatorio, paradiso, e numeri romani)
# Aggiungiamo un gruppo di cattura per separare i titoli di canto
canti = re.split(r'(?<=\n)(inferno|purgatorio|paradiso):\s*canto\s*[ivxlcdm]+\n*', text)

# Rimuoviamo gli eventuali vuoti dalla lista (risultati vuoti tra i titoli dei canti)
# Inoltre, rimuoviamo i titoli di canto non necessari
canti = [canto.strip() for canto in canti if canto.strip() and canto.lower() not in ['inferno', 'purgatorio', 'paradiso']]

if canti[0].lower() in ['inferno', 'purgatorio', 'paradiso']:
    canti = canti[1:]

print(canti)
print(len(canti))
print(canti[6])
```
